1. PANDAS (25 points)

```python
# 1 (2 marks). Load the netflix_titles.csv dataset as a Pandas dataframe:
import pandas as pd

netflix = pd.read_csv('data/netflix_titles.csv')
```

```python
# 2 (2 marks). Show the first 10 rows of the dataset
netflix.head(10)
```

```python
# 3 (1 mark). Show the last 5 rows
netflix.tail(5)
```

```python
# 4 (4 marks). Count the number of rows which have no missing value in any of the columns, and show the resulting data frame.

netflix_no_missing = netflix.dropna()

# Count the number of rows without missing values
num_rows_no_missing = netflix_no_missing.shape[0]

print("\nNumber of rows without missing values:", num_rows_no_missing)

```

```python
# 5 (3 marks). Drop everything that is not a 'Movie' and show the resulting data frame.
#This new dataframe, netflix_movies, which contains only the rows where the 'type' column is equal to 'Movie'. 
netflix_movie = netflix.loc[netflix['type'] == 'Movie']

# Display the resulting DataFrame
print("DataFrame with only 'Movie' entries:")
print(netflix_movie)

# Check the number of missing values in 'Duration' for movies
missing_duration_count = netflix_movie['duration'].isnull().sum()

# Display the result
print("Number of missing values in 'Duration' for movies:", missing_duration_count)
```

```python
# 6 (7 marks). For every movie where the duration is missing, replace it with the average
  # duration of the movies in the dataset.

# Convert 'Duration' to numeric format (assuming it contains both numbers and strings)
netflix_movie['duration'] = pd.to_numeric(netflix_movie['duration'], errors='coerce')

# Calculate the average duration of movies in the dataset
average_duration_movies = netflix_movie['duration'].mean(skipna=True)

# Display the number of missing values in 'Duration' for movies before replacement
print("Number of missing values in 'Duration' for movies before replacement:", netflix_movie.loc[netflix_movie['duration'].isnull(), 'duration'].shape[0])

# Replace missing values in 'Duration' for movies with the calculated average
netflix_movie.loc[netflix_movie['duration'].isnull(), 'duration'] = average_duration_movies

# Display the number of missing values in 'Duration' for movies after replacement
print("Number of missing values in 'Duration' for movies after replacement:", netflix_movie.loc[netflix_movie['duration'].isnull(), 'duration'].shape[0])

# Display the resulting DataFrame
print("DataFrame with missing 'Duration' values for movies replaced:")
print(netflix_movie)
```

```python
# 7 (3 marks). Are there any duplicates in the data? If yes, please remove them.
# Check for duplicates in the entire DataFrame

duplicates_exist = netflix_movie.duplicated().any()

# Display the result
if duplicates_exist:
    print("Duplicates exist. Removed duplicates. Updated DataFrame:")
else:
    print("No duplicates found. Original DataFrame:")

```

```python
# 8 (2 marks). Rename all columns to start with a capital letter.
# Rename all columns to start with a capital letter
netflix_movie.columns = netflix_movie.columns.str.capitalize()

# Display the DataFrame with updated column names
print(netflix_movie)
```

```python
# 9 (1 mark). Save the cleaned dataset as a new csv file.

# Save the cleaned dataset as a new CSV file
netflix_movie.to_csv('cleaned_netflix_dataset.csv', index=False)

# Display a message indicating that the file has been saved
print("Cleaned dataset saved as 'cleaned_netflix_dataset.csv'")
```

```python
# 1 (10 marks). Create a visualisation which shows the frequency of movies by year of release.
import matplotlib.pyplot as plt

# Convert 'Release_year' to numeric format (assuming it contains both numbers and strings)
netflix_movie['Release_year'] = pd.to_numeric(netflix_movie['Release_year'], errors='coerce')

# Filter movies and check their unique 'Release_year' values
movies = netflix_movie[netflix_movie['Type'] == 'Movie']
print("Unique Release Years for Movies:", movies['Release_year'].unique())

# Print movies with NaN in the 'Release_year' column
print("Movies with NaN in Release Year:")
print(movies[movies['Release_year'].isnull()])

# Check the counts of movies by year
movie_counts_by_year = movies['Release_year'].value_counts().sort_index()

# Plot the bar chart
plt.figure(figsize=(10, 6))
movie_counts_by_year.plot(kind='bar', color='skyblue')
plt.title('Frequency of Movies by Year of Release')
plt.xlabel('Year of Release')
plt.ylabel('Number of Movies')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

```python
# 2 (15 marks). Create a pie chart which shows the 5 top most popular genres of the movis
  # in the dataset.
import matplotlib.pyplot as plt


# Split the genres and explode the rows
genres_split = netflix['Genres'].str.split(',').explode()

# Count the frequency of each genre
top_genres = genres_split.value_counts().nlargest(5)

# Plot the pie chart
plt.figure(figsize=(8, 8))
top_genres.plot.pie(autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen', 'orange', 'lightpink'])
plt.title('Top 5 Most Popular Genres')
plt.show()

```
